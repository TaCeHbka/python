```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('kc_house_data.csv')
pd.set_option("display.float_format", "{:.2f}".format)
df.head(10)
```

```python
plt.hist(df['price']/1000)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость, тыс.')
plt.ylabel('Количество')
```

#### Вывод: 
Большая часть домов стоимостью менее 1 млн.

```python
plt.hist(df['sqft_living'])
plt.title('Распределение жилой площади домов')
plt.xlabel('Жилая площадь домов')
plt.ylabel('Количество')
```

#### Вывод: 
Более 10 тыс. домов имеет жилую площадь около 2000 кв.м.

```python
plt.hist(df['yr_built'])
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
```

#### Вывод: 
В среднем в продаже находятся дома, построенные в период с 1950 по 2000 года. Современных домов (год постройки от 2010) чуть больше среднего количества.

## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import seaborn as sns
sns.jointplot(x=df['sqft_living'],y=df['price']/1000,kind='reg');
```

#### Вывод: 
По оси х располагается первый признак - жилая площадь домов, а по оси у - второй признак - стоимость недвижимости в тыс. у.е. По середине точечный график их взаимного распределения. На диаграмме видна связь: чем больше жилая площадь дома, тем больше стоимость недвижимости. 


```python
sns.jointplot(x=df['grade'],y=df['price']/1000,kind='reg');
```

#### Вывод: 
По оси х располагается первый признак - оценка недвижимости, а по оси у - второй признак - стоимость недвижимости в тыс. у.е. По середине точечный график их взаимного распределения. На диаграмме видна связь: чем лучше оценка недвижимости, тем больше стоимость недвижимости.

```python
sns.jointplot(x=df['sqft_above'],y=df['price']/1000,kind='reg');
```

#### Вывод: 
По оси х располагается первый признак - площадь вокруг дома, земля вокруг, а по оси у - второй признак - стоимость недвижимости в тыс. у.е. По середине точечный график их взаимного распределения. На диаграмме видна связь: чем больше площадь вокруг дома, земли вокруг дома, тем больше стоимость недвижимости. 

```python
dfc=df.corr()
dfch=dfc.head(2)
dfch
```

```python
sns.heatmap(dfch, annot = True, linewidth=.5, cmap='coolwarm')
```

#### Вывод: 
По матрице корреляции можно увидеть признаки имеющие связь с признаком "стоимость недвижимости" = price.
Так, например, выше построенные графики по признакам "sqft_living", "grade", "sqft_above" наглядно показывают наличие линейной связи между ними и стоимостью недвижимости. А данная матрица показывает, что коэффициент корреляции (показывающий зависимость признков) по указанным признакам соответственно равен 0,70; 0,67; 0,61, что также подтверждает наличие связи между этими признаками и стоимостью недвижимости.
